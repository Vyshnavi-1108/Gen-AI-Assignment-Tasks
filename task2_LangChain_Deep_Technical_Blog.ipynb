7. Tool Example

```python
# Tool Example (Safe Calculator)

def calculator(input):
    try:
        return eval(input)
    except:
        return "Invalid expression"

tool = Tool(
    name="Calculator",
    func=calculator,
    description="Performs basic math calculations"
)

print("Tool Output:", tool.invoke("5+3"))
```
